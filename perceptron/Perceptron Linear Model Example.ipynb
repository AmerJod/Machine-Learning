```python
from perceptron import Perceptron, Point

import numpy as np
import matplotlib.pyplot as plt

# Define the linear function the data points should follow.
func = lambda x: 0.25 * x + 0.25

# Create data points
data_points = [Point(-1 + 2 * np.random.rand(1), np.random.rand(1), func) for i in range(0, 500)]
```

```python
# Scatter plot to see the two different classes of points
for data_point in data_points:
    if(data_point.label == 1):
        plt.scatter(data_point.x, data_point.y, color='r')
    else:
        plt.scatter(data_point.x, data_point.y, color='b')
plt.show()
```

```python
# Create the perceptron
p = Perceptron(2)

# Train the perceptron on the data points
for epoch in range(500):
    num_dp = len(data_points)
    point_indices = np.random.choice(num_dp, 100, replace=False)
    epoch_data = [data_points[i] for i in point_indices]
    
    p.store_errors(data_points)
    
    for data_point in epoch_data:
        p.train(data_point, 0.0001)
```

```python
# Draw the linear line the perceptron learns.
draw_points_x = [float(i)/10 for i in range(-11, 11)]
draw_points_y = map(lambda x: p.func_evaluate(x), draw_points_x)
plt.plot(draw_points_x, draw_points_y, color='g', linewidth=3)

for data_point in data_points:
    if(data_point.label == 1):
        plt.scatter(data_point.x, data_point.y, color='r')
    else:
        plt.scatter(data_point.x, data_point.y, color='b')
plt.show()
```

```python
# Plot the convergence curve of the perceptron's learning.
print("Number of errors after training (100 epochs)", p.errors[-1])
plt.plot(p.errors)
plt.xlabel("Number of epochs")
plt.ylabel("Number of errors")
plt.title("Number of errors vs epochs")
plt.show()
```
